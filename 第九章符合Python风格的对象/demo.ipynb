```python
# Python 内置函数 eval() 详解
# eval() 将一个字符串当作 Python 表达式来执行，并返回结果。

eval("3+2*8")
```

```python
eval("print('eval 输出 不安全！')")
```

```python
# 一些有趣的用法
def make_function(expr, var='x'):
    def f(x):
        return eval(expr)
    return f
```

```python
square = make_function("x ** 2")
square(5)
```

```python
cube = make_function("x ** 3")
cube(3)
```

```python
class Demo:
    def instance_method(self):      # Python 自动传入 实例
        pass

    @classmethod
    def class_method(*args):          # Python 自动传入 类
        return args

    @staticmethod
    def static_method():            # Python 什么都不自动传
        pass
```

```python
Demo.class_method() # Python 自动传入 类
```

```python
Demo.class_method("hello")
```
