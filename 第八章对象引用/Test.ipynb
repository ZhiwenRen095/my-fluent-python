```python
class Bank:
    bank_name = "Python银行"

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            print("存款金额必须为正数")
            return
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"余额不足，当前余额: {self.balance}")
            return
        self.balance -= amount

    @classmethod
    def from_string(cls, s: str):
        owner, balance = s.split("-")
        return cls(owner, int(balance))

    def __str__(self):
        return f"户主: {self.owner}, 余额: {self.balance}"
```

```python
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age     # 这里会触发 setter

    @property              # getter：读取时调用
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):    # setter：赋值时调用
        if new_age < 0:
            raise ValueError("Age must be greater than 0!")
        self._age = new_age
```
