```python
def gen_123():
    yield 1
    yield 2
    yield 3

gen_123
```

```python
gen_123()

for num in gen_123():
    print(num)
```

```python
def gen_AB():
    print('start')
    yield 'A'

    print('continue')
    yield 'B'

    print('end')

res1 = [x*3 for x in gen_AB()]
```

```python
for i in res1:
    print(i)
```

```python
res2 = (x*3 for x in gen_AB())
res2
```

```python
for x in res2:
    print("--->" + x)
```

```python
def chain(*iterables):
    for it in iterables:
        for i in it:
            yield i

s = 'ABC'
t = tuple(range(3))

list(chain(s, t))
```

```python
def chain(*iterables):
    for i in iterables:
        yield from i

s = 'ABC'
t = tuple(range(3))

list(chain(s, t))
```

```python
def averager():
    total = 0.0
    count = 0
    average = None
    while True:
        # 1. 挂起并产出当前的 average
        # 2. 接收外部通过 .send() 传进来的 term
        term = yield average
        if term is None:
            break
        total += term
        count += 1
        average = total / count

# 使用示例
coro_avg = averager()
next(coro_avg)      # 预激 (Prime)：运行到第一个 yield，此时返回 None

print(coro_avg.send(10))  # 发送 10，返回 10.0
print(coro_avg.send(20))  # 发送 20，返回 15.0
print(coro_avg.send(30))  # 发送 30，返回 20.0
```
