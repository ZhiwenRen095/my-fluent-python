```python
from dataclasses import dataclass


@dataclass
class Piece:
    name: str
    value: int


pieces = [
    Piece("Pawn", 1),
    Piece("Rook", 5),
    Piece("Bishop", 3),
    Piece("Knight", 3),
    Piece("GoldGeneral", 5)
]

sorted(pieces, key=lambda x: (x.value * -1, len(x.name)))
```

```python
words = ["banana", "pie", "apple", "fig", "cherry", "kiwi", "date"]

# 先按长度升序，长度相同按字母序
# 期望输出: ['fig', 'pie', 'date', 'kiwi', 'apple', 'banana', 'cherry']
sorted(words, key=lambda x: (len(x), x))
```

```python
tasks = [
    {"name": "deploy", "priority": "HIGH"},
    {"name": "test", "priority": "MEDIUM"},
    {"name": "hotfix", "priority": "CRITICAL"},
    {"name": "refactor", "priority": "LOW"},
    {"name": "bugfix", "priority": "HIGH"},
    {"name": "doc", "priority": "LOW"},
]

# 按优先级排序：CRITICAL > HIGH > MEDIUM > LOW，同优先级按 name 字母序
# 期望输出: hotfix → bugfix → deploy → test → doc → refactor
priority_order = {
    "LOW": 0,
    "MEDIUM": 1,
    "HIGH": 2,
    "CRITICAL": 3
}
sorted(tasks, key=lambda x: (priority_order[x["priority"]] * -1, x["name"]))
```

```python
items = ["orange", "apple", "★ pinned", "banana", "★ important", "grape"]

# 以 ★ 开头的元素排在最前面（保持它们之间的原始顺序），其余按字母序
# 期望输出: ['★ pinned', '★ important', 'apple', 'banana', 'grape', 'orange']
sorted(items, key=lambda x: (x.startswith("★") * -1, x))
```

```python
ips = ["192.168.1.10", "10.0.0.1", "192.168.1.2", "172.16.0.255", "10.0.0.12"]

# 按 IP 地址的数值大小排序
# 期望输出: ['10.0.0.1', '10.0.0.12', '172.16.0.255', '192.168.1.2', '192.168.1.10']
sorted(ips, key=lambda x: tuple(int(y) for y in x.split(".")))
```

```python
data = [5, "banana", 3, "apple", 8, "cherry", 1]

# 数字排前面（升序），字符串排后面（字母序），不能报 TypeError
# 期望输出: [1, 3, 5, 8, 'apple', 'banana', 'cherry']
sorted(data, key=lambda x: (isinstance(x, int) * -1, x))
```

```python
stream = ['b', 'a', 'c', 'a', 'b', 'd', 'c', 'e']

# 去重，并按每个元素最后一次出现的位置排序
# 'a' 最后在 index 3, 'b' 最后在 index 4, 'c' 最后在 index 6, 'd' 在 5, 'e' 在 7
# 期望输出: ['a', 'b', 'd', 'c', 'e']
last_index = {value: index for index, value in enumerate(stream)}
sorted(last_index, key=lambda x: last_index[x])
```

```python
import time
from functools import wraps

def timer(func):
    """记录函数执行时间的装饰器"""
    @wraps(func)  # 保持原函数的元数据（如函数名、注释文档等）
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()

        # 执行被装饰的函数
        result = func(*args, **kwargs)

        end_time = time.perf_counter()
        print(f"函数 [{func.__name__}] 运行耗时: {end_time - start_time:.6f} 秒")

        return result
    return wrapper
```

```python
needles = {"abc","ahjsdcbajs",123}
haystack = {123}
```

```python
@timer
def test1(n):
    for i in range(n):
        found = len(needles & haystack)
```

```python
@timer
def test2(n):
    for i in range(n):
        found = 0
        for n in needles:
            if n in haystack:
                found += 1
```

```python
epoch = 10**8
test1(epoch)
```

```python
test2(epoch)
```
