```python
def factorial(n:int):
    """return n!"""
    return 1 if n < 2 else n * factorial(n-1)

factorial(42)
```

```python
# 在 Python 的世界里，“一切皆对象”这句话的完整表达应该是：“一切皆一等对象”。
print("函数是一等对象")
factorial.__doc__
```
