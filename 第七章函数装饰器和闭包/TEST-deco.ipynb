```python
import time
from functools import wraps
```

```python
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"[{func.__name__}]函数执行{end - start:.4f}s。")
        return result

    return wrapper
```

```python
@timer
def add(a, b):
    time.sleep(1)
    return a + b
```

```python
add(19, 21)
```

```python
def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for i in range(n):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator
```

```python
@repeat(10)
def print_name(name):
    print(name)
```

```python
print_name("rzw")
```

```python
current_user = {"name": "Alice", "role": "user"}

def require_role(role):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            if current_user.get("role") != role:
                print(f"权限不足！需要 '{role}' 角色，当前角色为 '{current_user.get('role')}'")
                return None
            return func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@require_role("admin")
def delete_database():
    print("数据库已删除")

@require_role("user")
def view_database():
    print("已成功查看用户资料~")
```

```python
delete_database()
view_database()
```

```python
def memoize(func):
    cache = {}
    @wraps(func)
    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]
    return wrapper
```

```python
@memoize
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(50)
```
